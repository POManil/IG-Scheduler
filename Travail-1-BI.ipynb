### Lignes
Chaque ligne contient les données d'un pokémon en fonction de son identifiant dans le pokédex national.
Les id sont compris entre 1 et 1017.

### Colonnes
Chaque colonne contient les données du pokémon dont il est question. 

#### Description des colonnes 
* id - Identifiant dans le pokédex national (numéro unique correspondant au pokémon)
* name - Nom du pokémon
* rank - Niveau de rareté (ou est bébé)
* generation - Génération de première apparition
* evolves_from - Pré-évolution
* type1 - Premier type
* type2 - Second type
* hp - Nombre de points de vie
* atk - Nombre de points d'attaque physique
* def - Nombre de points de défense
* spatk - Nombre de points d'attaque spéciale
* spdef - Nombre de points de défense spéciale
* speed - Nombre de points de vitesse
* total - Somme des points d'attributs (hp, atk, def, ..., speed)
* height - Taille (en décimètres) 
* weight - Poids (en centaines de grammes)
* abilities - Talent(s) du pokémon
* desc - Description du pokémon dans le pokédex
